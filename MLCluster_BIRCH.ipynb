```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import Birch
brh = Birch(threshold=1.5,branching_factor = 50, n_clusters=3)
brh.fit(x)

```

```python
labels=brh.labels_

```

```python
labels
```

```python
supervised=dataset
supervised['Cluster'] = labels
```

```python
supervised
```

```python
import seaborn as sns
facet=sns.pairplot(data=supervised,vars=supervised.iloc[:,[3,4]],hue=supervised.columns[5],palette='viridis', 
                   markers='o', height=2.5, aspect=1.2
            )
```

```python
y_brh=brh.fit_predict(x)
```
